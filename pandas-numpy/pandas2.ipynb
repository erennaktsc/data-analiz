```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
# Soru : Ödenen ücret 40 'dan büyük olan satırları getirin.
```

```python
df[df['total_bill'] >40]
```

```python
# Soru: Cinsiyeti Erkek olan satılar getirin.
```

```python
df[df['sex'] == 'Male']
```

```python
#Soru : Cinsiyeti kadın ve bahşiş miktarı 5'den büyük olan satıları getirin.
```

```python
df.head()
```

```python
df[(df['sex'] == 'Female') & (df['tip']> 5)] 
# & ve sembolü ayrıca sorguları parantezde göstermek ZORUNDAYIZ
```

```python
# Soru : Günü Pazar(Sun) olan ve ödenen miktar 30'dan büyük olan satırları getirin.
```

```python
df.head()
```

```python
df[(df['day'] == 'Sun') & (df['total_bill'] >30)]
```

```python
# Soru : Cumartesi(Sat) VEYA(|) Pazar(Sun) günü olan satırları getirin.
```

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')] # | işareti veya işareti
```

```python
# Soru : Cumartesi(Sat) veya Pazar(Sun) veya Perşembe(Thur) günü olan satırları getirin.
```

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun') | (df['day'] == 'Thur')]
```

```python
df[df['day'].isin(['Sat' ,'Sun' , 'Thur'])]  #ÜSTDEKİ KOD YERİNE BUNU KULLANN!!!
```
