### Soru : Numpy kullanarak cos(x) ve sin(x) grafiklerini çizdilerim.
### cos(x) plt.scatter, sin(x) plt.plot olsun


```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x,y2, color = "red", label = "Cosx"  )
plt.plot(x,y1,color = "green", label = "Sinx")
plt.legend()
plt.show()
```

## plt.plot ve plt.scatter Parametreleri

```python
plt.plot(x,y1,color = "green", label = "Sinx", linestyle="--") # linestyle çizgi stilini belirler
plt.legend()
plt.show()
```

```python
x =[0 ,1 , 2, 3, 4, 5, 6]
y =[0 ,2 , 4, 6, 8, 10,12]

plt.figure(figsize=(6,4)) #Çıktı boyutunu değiştiririz.figsize=(en,yükseklik)

plt.plot(x,y,color="magenta", marker = "o" , markersize =20, alpha = 0.2,
        linewidth=10)
#alpha opaklığını-saydamlığını değiştirmek için kullanılır
plt.show()
```

```python
x =[0 ,1 , 2, 3, 4, 5, 6]
y =[0 ,2 , 4, 6, 8, 10,12]

plt.plot(x, y, marker='o', mfc='yellow', mec='black', mew=2, alpha=0.8)
# mec = markeredgecolor	(İşaretçi kenar rengi)
# mew = markeredgewidth (İşaretçi kenar kalınlığı)
plt.show()
```

```python

# X verisi: 0'dan 10'a kadar 100 nokta
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Grafik çizimi
plt.plot(x, y, label='sin(x)', color='blue')

# x ve y eksen sınırlarını ayarla
plt.xlim(2, 8)   # X ekseni 2 ile 8 arasında gösterilsin
plt.ylim(-1.5, 1.5)  # Y ekseni -1.5 ile 1.5 arası

# Eksen üzerindeki değerleri özelleştir
plt.xticks([2, 4, 6, 8], labels=['iki', 'dört', 'altı', 'sekiz'])
plt.yticks([-1, 0, 1])

# Diğer ayarlar
plt.title("Sinüs Grafiği (xlim, xticks örneği)")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.grid(True)
plt.legend()
plt.show()
```

## plt.scatter

```python
x = np.arange(100) / 10
plt.scatter(x , np.cos(x) , color = "red",
              label = "Cosx" ,
                s=40, edgecolors="blue" , linewidths=40,facecolor="black")

# s siyah noktanın  kalınlığını ve mavi çemberin  kalınlığını belirler
# edgecolors ile siyah yuvarlakların çevresini mavi ile belirledik
# linewidths ile siyah yuvarlakların etrafını saran maviliklerin boyutunu ayarladık
plt.legend()
plt.show()
```

# plt.pie()

```python
# Pasta grafiği (pie chart), verileri dilimlere bölerek oranları göstermeye yarar.
# Bütünün parçalarını karşılaştırmak için kullanılır.
# Örneğin, bir şirketin satış oranlarını, bir sınavın başarı yüzdelerini veya bir şehrin nüfus dağılımını göstermek için uygundur.
```

```python
degerler = [30, 20, 50]  # Yüzdeler toplamı 100 olmak zorunda değil.
kategoriler = ["A Kategorisi", "B Kategorisi", "C Kategorisi"]

plt.pie(degerler,labels=kategoriler, autopct="%1.1f%%")
#  autopct="%1.1f%%" yüzdeleri otomatik hesaplayıp verir.
plt.title("Temel Pasta Grafiği")
plt.show()
```

```python
plt.pie(degerler,labels=kategoriler, 
        autopct="%1.1f%%",colors=["orange","blue" ,"purple"],
        radius=1.4)
# radius ile çıktıyı büyüttük
plt.show()
```

```python
plt.pie(degerler, labels=kategoriler, autopct="%1.1f%%",
        shadow=True) # shadow=True ile gölgelendirme yaptık.
plt.show()
```

```python
plt.pie(degerler, labels=kategoriler, autopct="%1.1f%%",
        shadow=True,startangle=90) # shadow=True ile gölgelendirme yaptık.
plt.show()
```
