```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Ne Zaman sns.swarmplot()?
Noktaların asla üst üste binmemesini ve her bir gözlemin tam yerini görmek istiyorsan

Veri seti nispeten küçük (ör. birkaç yüz–bin satır) olduğunda performans sorunu yaşamazsın

Alt grupları (hue) bir arada, ama çakışma olmadan yan yana göstermek istiyorsan

Örneğin, 300–500 kayıttan oluşan bir kategorik dağılımda tüm noktaları netçe ayırmak istediğinde

1. Ortak Parametreler

Parametre	Ne İçin?	Örnek Kull.
x, y	Hangi değişkenlerin eksenlerde olacağını belirler.	x="day", y="total_bill"
data	Veri çerçevesi (DataFrame)	data=tips
hue	Alt grupları renklerle ayırır	hue="sex"
dodge	hue varsa alt grupları yanyana mı üst üste mi çizsin	dodge=True (yan yana)
palette	Renk paletini belirler (hue için)	palette="pastel"
order	Kategori sırasını kontrol eder	order=["Thur","Fri","Sat","Sun"]
hue_order	hue alt gruplarının sıralamasını belirler	hue_order=["Male","Female"]
orient	Grafiği yatay mı dikey mi çizsin ("h"/"v")	orient="h"
size	Nokta boyutunu ayarlar	size=6
alpha	Nokta saydamlığını ayarlar	alpha=0.7
linewidth	Nokta kenar çizgi kalınlığı	linewidth=1

```python
data = pd.read_csv("supermarket_sales.csv")
data.head()
```

```python
sns.swarmplot(data=data , y='Total')
```

```python
sns.swarmplot(data=data , x='Payment' , y='Total', palette="pastel") 
```

```python
sns.swarmplot(data=data , x="Payment" , y='Total' , hue='Gender')
```

```python
sns.swarmplot(data=data , x='Payment' , y='Total' , hue='Gender' , dodge=True)
```

```python
sns.swarmplot(data=data , x='Payment' , y='Total' , hue='Gender' ,
               color="gold" ,edgecolor="black", size=5)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=data , x='Payment' , y='Total' )
sns.swarmplot(data= data  , x='Payment' , y='Total',color='black' )
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=data , x='Payment' , y='Total',inner=None,width=0.5)
sns.swarmplot(data= data  , x='Payment' , y='Total',color='white' )
plt.show()
```
