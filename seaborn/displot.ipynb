```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

sns.displot(), dağılım (distribution) grafikleri çizmek için kullanılan çok amaçlı bir fonksiyondur.

Özellikle sayısal verilerin dağılımını görmek için kullanılır.

Tek bir fonksiyonla:

Histogram (kind="hist")

KDE (Yoğunluk Grafiği) (kind="kde")

ECDF Grafiği (kind="ecdf") çizebilirsin.

Yani displot = dağılım çizim merkezi gibi düşün!

Parametre | Ne işe yarar? | Örnek

x= | Hangi değişkeni çizeceğiz | x="age"

data= | Veri seti | data=df

kind= | Grafik türü seçimi | "hist", "kde", "ecdf"

hue= | Farklı gruplar için farklı renkler | hue="gender"

col=, row= | Alt grafiklere ayırmak | col="species"

binwidth= | Histogramda her barın genişliği | binwidth=5

multiple= | Üst üste mi yoksa yan yana mı gösterecek | "stack", "dodge", 
"layer", "fill"

```python
data = pd.read_csv("supermarket_sales.csv")
data = data[['Gender' , 'Payment' , 'Unit price' , 'Quantity' , 'Total' , 'gross income','Branch']]
```

```python
sns.displot(data=data , x='Total' , kde=True)
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True)
# rug=True yaparsan, grafiğin altına küçük çizgiler ekler.
# Her çizgi, bir veri noktasını temsil eder.
# Böylece veri yoğunluğunu ham veri seviyesinde görmüş olursun.
# Rug çizgileri, verilerin tam olarak nerelerde yoğunlaştığını gösterir.
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True,rug_kws={'height': 0.05})
# rug_kws, rug çizgilerinin (küçük çubukların) stilini özelleştirmeni sağlar.
# Rengini (color)
# Kalınlığını (linewidth)
# Şeffaflığını (alpha)
# değiştirebilirsin.
# "rug" çizgilerinin görünüm ayarları için kullanılır.
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True , rug_kws={'height':0.05},
            kde_kws={'bw_adjust':0.6})
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True , rug_kws={'height':0.05},
            hue='Payment')
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True , rug_kws={'height':0.05},
            hue='Payment' , multiple='stack')
plt.show()
```

```python
sns.displot(data=data , x='Total', kde=True , rug=True , rug_kws={'height':0.05},
            hue='Payment' , multiple='stack' , col='Gender')
plt.show()
```

```python
sns.displot(data=data , x='Total' , kde=True , rug=True , rug_kws={'height':0.05},
            hue='Payment' , multiple='stack' , col='Gender',
            row='Branch')
plt.show()
```

```python
sns.displot(data=data , x='Total' , kind='ecdf',
            hue='Payment' , col='Gender',
            row='Branch')
plt.show()
```
