```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

relplot, Seaborn kütüphanesindeki "relational plot" (ilişkisel grafik) anlamına gelir.

Aslında bir “fabrikadır”: İçinde hem scatterplot hem de lineplot çalıştırabilir.

Çoklu boyutlarda, farklı grupları, alt grafiklerle (facet) birlikte çizmek için kullanılır.

```python
data = sns.load_dataset('flights')
```

```python
data.head()
```

```python
sns.relplot(data=data , x='year' , y='passengers',s=100,
            hue='passengers')
plt.show()
```

```python
sns.relplot(data=data , x='year' , y='passengers',kind='line')
plt.show()
```

```python
february_data = data.query("month == 'Feb'")
sns.lineplot(data=february_data , x='year', y='passengers')
# burada sadece şubat ayının verilerinin grafiğini çizdik
plt.show()
```

```python
sns.lineplot(data=data ,x='year' , y='passengers',hue='month')
plt.show()
```

```python
sns.lineplot(data=data , x='year' , y='passengers',hue='month',
             style='month')
plt.show()
```

```python
sns.relplot(data=data , x='year' , y='passengers', hue='month',
            kind='line')
plt.title('Number of Passengers Over Years')
plt.show()
```
