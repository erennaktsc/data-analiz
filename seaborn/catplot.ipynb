```python
data = pd.read_csv("supermarket_sales.csv")
data.head()
```

```python
sns.catplot(data=data , x="Payment" , y='Total'
             ,kind='box',
            palette='pastel')
# kind= hangi tÃ¼rde grafik Ã§izeceÄŸimizi belirler
plt.show()
```

```python
sns.catplot(data=data , x='Payment' , y='Total' ,hue='Gender' 
            ,  kind='violin')
plt.show()
```

```python
sns.catplot(data=data , x='Payment' , y='Total' ,
            col='Gender' , kind='violin')
# col= veriyi sÃ¼tunlara (yan yana grafiklere ayÄ±rÄ±r)
plt.show()
```

```python
sns.catplot(data=data , x='Payment' , y='Total',
            col='Gender' , kind='violin' ,row='Customer type' , palette='pastel')
# row= Veriyi satÄ±rlara (alt alta grafikleri) ayÄ±rÄ±r.
plt.show()
```
