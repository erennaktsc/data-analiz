```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

sns.ecdfplot(), bir ECDF grafiği (Empirical Cumulative Distribution Function) çizer.

ECDF, verinin dağılımını kümülatif (birikimli) olarak gösterir.

Yani:

"Bir değerin o değere kadar olan gözlemlerde yüzde kaç sıklıkta gerçekleştiğini" gösterir.

ECDF, Histogramdan daha hassas bir analiz sunar.

Çok küçük veri setlerinde bile güzel çalışır (histogram bazen veri azsa kötü görünür).

```python
data = pd.read_csv("supermarket_sales.csv")
data = data[['Gender' , 'Payment' , 'Unit price' , 'Quantity' , 'Total' , 'gross income']]
```

```python
sns.ecdfplot(data=data , x='gross income')
plt.show()
```

```python
sns.ecdfplot(data=data , x='gross income' , hue='Gender')
plt.show()
```

```python
sns.ecdfplot(data=data , x='gross income' , hue='Gender' , stat='count')
plt.show()
```

```python
sns.ecdfplot(data=data , y='gross income' , hue='Gender' , stat='count')
plt.show()
```
