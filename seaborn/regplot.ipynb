```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

regplot, Seaborn kütüphanesindeki “regression plot”, yani doğrusal regresyon (lineer ilişki) grafiğidir.

İki sayısal değişken arasındaki ilişkiyi hem scatter plot (dağılım) hem de üzerine bir regresyon çizgisi ekleyerek gösterir.

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.regplot(data=data ,x='total_bill' , y='tip')
plt.show()
```

```python
sns.regplot(data=data , x='total_bill' , y='tip',color='orange',
            marker='*')
plt.show()
```

```python
sns.regplot(data=data , x='total_bill',y='tip',
            color='orange',marker='*',
            line_kws=dict(color='green' , linestyle='--'))
plt.show()
```

```python
sns.regplot(data=data ,x='total_bill' , y='tip',
            # color='orange',marker='*',
            line_kws=dict(color='green' , linestyle='--'),
            scatter_kws=dict(s=50 , color="orange" , alpha=0.3))
plt.show()
```

```python
sns.regplot(data=data , x='total_bill' , y='tip',
            # color='orange',marker='*',
            line_kws=dict(color='green' , linestyle='--'),
            scatter_kws=dict(s=50 , color='orange',alpha=0.5),
            ci=None)
plt.show()
```

```python
sns.regplot(data=data , x='total_bill' ,y='tip',
            color='orange' , marker='*',
            line_kws=dict(color='green' , linestyle='--'),
            scatter_kws=dict(s=100, color='orange', alpha = 0.4),
            ci=50 , n_boot=500 )
# n_boot, Seaborn grafiklerinde (özellikle lineplot, barplot, pointplot) güven aralığı (confidence interval) çizilirken kullanılan bir parametredir.
# Bootstrap = Veriden tekrar tekrar küçük örnekler alıp ortalamalarını alarak güvenli bölgeyi tahmin etmek demektir.
plt.show()
```
