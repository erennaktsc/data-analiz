```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

clustermap(), Seaborn kütüphanesinde yer alan bir fonksiyondur.

Verideki benzerliklere göre satır ve/veya sütunları otomatik olarak gruplayan (cluster eden) ve bunu ısı haritası (heatmap) şeklinde görselleştiren bir araçtır.

Temel farkı:

Sadece heatmap çizmez, aynı zamanda veri üzerinde hiyerarşik kümeleme (clustering) uygular.

```python
data = pd.read_excel('mart_linePlot.xlsx')
data.columns = data.columns.str.lower()
```

```python
DataPiv = data.pivot_table(index='outlet_year' , columns='outlet_size' , values='sales')
DataPiv
```

```python
sns.clustermap(DataPiv)
plt.show()
```

```python
# Clustermap çizimi
sns.clustermap(
    DataPiv,
    cmap="YlGnBu",             # Renk skalası: Açık sarı → koyu mavi
    annot=True,                # Hücre değerlerini göster
    fmt=".0f",                 # Sayıları tam sayı olarak göster
    linewidths=0.5,            # Hücreler arası çizgi
    linecolor='white',         # Çizgi rengi beyaz
    standard_scale=1           # Sütunları normalize et (ölçek farklarını kaldır)
)

plt.title("Yıllara ve Mağaza Tipine Göre Satış Clustermap")
plt.show()
```

Grafik Nasıl Yorumlanır?
✅ Sol Dendrogram (Satırlar)
Yıllar benzer satış desenine göre gruplandı

Örneğin 1997 ve 1999 yan yana gelirse → benzer satış profiline sahip

✅ Üst Dendrogram (Sütunlar)
High, Medium, Small outlet tipleri satış ortalamalarına göre gruplanır

Örneğin High ve Small aynı kümeye girerse → benzer davranış gösteriyor olabilir

✅ Renkler
Koyu mavi = yüksek satış

Açık sarı = düşük satış
→ Hangi yıl ve mağaza tipi daha çok satmış anında görünüyor
