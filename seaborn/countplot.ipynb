```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("supermarket_sales.csv")
```

```python
df.head()
```

```python
df.drop("Branch" , axis=1,inplace=True)
```

```python
df.head()
```

# Sütun grafiği, kategorik değişkenleri görselleştirmede kullanılır.


```python
plt.figure(figsize=(15,5))
sns.countplot(df,x="Product line",color="purple",
              edgecolor="black",width=0.3,linewidth=5)
plt.show()
```

```python
sns.countplot(data=df,y="Product line")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x="Product line", hue="Gender")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x="Product line", hue="Gender",palette="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df, x='Product line', 
              color="purple", linewidth=2, edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='Product line',
              facecolor=(1,0,1,0),linewidth=5,
              edgecolor=sns.color_palette('dark',3))
#✅ facecolor=(1,0,1,0) → Çubukların iç rengini belirler.
# edgecolor=sns.color_palette('dark',3) → Çubukların kenar rengini belirler.
plt.show()
```
