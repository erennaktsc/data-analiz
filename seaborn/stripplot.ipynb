```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Ne Zaman sns.stripplot()?
Büyük veri setlerinde noktaların üst üste binmesini önlemek için hafif bir dağıtma (jitter) istiyorsan

Hız önemliyse, çünkü jitter uygulamak swarm algoritmasına göre çok daha hızlıdır

Nokta yoğunluğunu kabaca görmek yeterliyse (ayrı ayrı tam konum gerekmez)



1. Ortak Parametreler

Parametre	Ne İçin?	Örnek Kull.
x, y	Hangi değişkenlerin eksenlerde olacağını belirler.	x="day", y="total_bill"
data	Veri çerçevesi (DataFrame)	data=tips
hue	Alt grupları renklerle ayırır	hue="sex"
dodge	hue varsa alt grupları yanyana mı üst üste mi çizsin	dodge=True (yan yana)
palette	Renk paletini belirler (hue için)	palette="pastel"
order	Kategori sırasını kontrol eder	order=["Thur","Fri","Sat","Sun"]
hue_order	hue alt gruplarının sıralamasını belirler	hue_order=["Male","Female"]
orient	Grafiği yatay mı dikey mi çizsin ("h"/"v")	orient="h"
size	Nokta boyutunu ayarlar	size=6
alpha	Nokta saydamlığını ayarlar	alpha=0.7
linewidth	Nokta kenar çizgi kalınlığı	linewidth=1

```python
data = pd.read_csv("supermarket_sales.csv")
data.head()
```

```python
sns.stripplot(data=data , y='Total')
plt.show()
```

```python
sns.stripplot(data=data , y='Total' , jitter= 0.2)
plt.show()
```

```python
sns.stripplot(data=data , y='Total' , jitter=0.2 , linewidth=0.3)
```

```python
sns.stripplot(data=data , x='Payment' , y='Total' , hue='Gender' ,
               jitter=0.2 , linewidth=0.3)
plt.show()
```

```python
sns.stripplot(data=data , x='Payment' , y='Total' , hue='Gender',
              jitter=0.2 , linewidth=0.3 , dodge=True )
```

```python
sns.stripplot(data=data , x='Payment' , y='Total' , jitter=0.1)
sns.violinplot(data=data , x='Payment' , y="Total" , color="gray")
```

```python
sns.stripplot(data=data , x='Payment' , y='Total' , color='black' , size=5 , palette="pastel")
sns.boxplot(data=data , x='Payment' , y='Total' , color="gold")
# palette	Renk paletini belirler (hue için)	palette="pastel"
# order	Kategori sırasını kontrol eder	order=["Thur","Fri","Sat","Sun"]
# hue_order	hue alt gruplarının sıralamasını belirler	hue_order=["Male","Female"]
```
