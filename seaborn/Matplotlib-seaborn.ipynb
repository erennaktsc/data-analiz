# Seabron ile Scatterplot

```python
# 🔹 sns.scatterplot() fonksiyonu, Seaborn kütüphanesinde kullanılan dağılım grafiği (scatter plot) çizmek için kullanılır.
# 🔹 İki değişken arasındaki ilişkiyi görmek için kullanılır.
# 🔹 Veri noktaları arasındaki eğilimleri veya kümelenmeleri analiz etmek için oldukça faydalıdır.
```

```python
data
```

```python
#! DİKKAT SAYISAL DEĞER OLAN SÜTUNLARI VERDİK
plt.figure(figsize=(10,6))
sns.scatterplot(data=data,
                x = "bill_depth_mm",
                y="flipper_length_mm",
                hue= "species")
plt.title("bill_depth_mm vs flipper_length_mm ")
plt.show()
# Eğer bir değişkeni kategoriye göre renklendirmek istersen, hue parametresini kullanabilirsin.
```

Bu scatter plot (saçılım grafiği), penguen türleri (species) arasındaki "bill_depth_mm" (gaga derinliği) ve "flipper_length_mm" (yüzgeç uzunluğu) ilişkisinin nasıl dağıldığını gösterir.

📌 Grafiği Yorumlama:
Her bir nokta bir pengueni temsil eder.
X ekseni: Gaga derinliği (bill_depth_mm)
Y ekseni: Yüzgeç uzunluğu (flipper_length_mm)
Renkler (hue="species")
Mavi (Adelie)
Turuncu (Chinstrap)
Yeşil (Gentoo)
🎯 Gözlemler:
✅ Gentoo penguenleri (Yeşil noktalar):

En uzun yüzgeçlere sahipler (210-230 mm)
Daha küçük gaga derinliğine sahipler (14-16 mm)
Üst kısımda kümelenmişler.
✅ Adelie (Mavi noktalar) ve Chinstrap (Turuncu noktalar) penguenleri:

Gaga derinliği daha fazla (17-21 mm)
Yüzgeç uzunlukları daha kısa (180-200 mm)
Birbirlerine yakın konumlanmışlar.
🔹 Sonuçlar ve Çıkarımlar:
Gentoo penguenleri, diğer türlerden bariz şekilde ayrılmış görünüyor.
Adelie ve Chinstrap penguenleri, yüzgeç uzunluğu açısından benzer ancak gaga derinliği açısından farklılık gösterebilir.
Eğer bir sınıflandırma modeli geliştirilecekse, Gentoo penguenleri kolayca ayrıştırılabilir, ancak Adelie ve Chinstrap'ı ayırmak biraz daha zor olabilir.

```python
sns.barplot(x=group_data.index, 
            y=group_data.values)
plt.show()
```

```python
cmap = plt.get_cmap("viridis")
```

```python
group_data.min()
```

```python
plt.Normalize() # 0-1 arasında renk sklasına normalize eder.
```

```python
norm = plt.Normalize(group_data.min() , group_data.max()) # 0 ile 1 arasında renk skalasını normalize et
```

```python
colors = cmap(norm(group_data.values))  # renkleri oluştur
colors
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
sns.barplot(x=group_data.index, 
            y=group_data.values,palette=colors)

ax.set_ylabel("Averge Body_Mass_g")
ax.set_xlabel("Species")
fig.add_axes([0.95, 0.1, 0.05, 0.8])
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
# fig.colorbar(sm)  BUNU ARAŞTIR
plt.show()
```
